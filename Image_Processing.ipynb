```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#scaling of image

img = cv2.imread(r'/content/LionKing.png')
cv2_imshow(img)
```

```python
(height,width) = img.shape[:2]
```

```python
res1 = cv2.resize(img,(int(width/2),int(height/2)),interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imwrite('result1.png',res1)
```

```python
cv2_imshow(res1)
```

```python
#ROTATION of image#
```

```python
(rows,cols) = img.shape[:2]
```

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)#((x,y),angle,scaling)
```

```python
res2 = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2_imshow(res2)
```

```python
#Translation of image
# Create translation matrix.
# If the shift is (x, y) then matrix would be
# M = [1 0 x]
#     [0 1 y]
# Let's shift by (300, 150).
```

```python
M1 = np.float32([[1,0,300],[0,1,150]])
```

```python
res3 = cv2.warpAffine(img,M1,(cols,rows))
```

```python
cv2_imshow(res3)
```
